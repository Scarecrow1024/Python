```python
import pandas as pd
import numpy as np
import pylab
%pylab
names = pd.read_csv('C:/Users/Janfer/Desktop/pydata-book-master/ch02/names/yob1880.txt', names=['name', 'sex', 'births'])
names.groupby('sex').births.sum()
names['sex'].value_counts()

years = range(1880, 2011)
pieces = []
cloumns = ['name', 'sex', 'births']
for year in years:
    path = 'C:/Users/Janfer/Desktop/pydata-book-master/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=cloumns)
    frame['year'] = year
    pieces.append(frame)
names = pd.concat(pieces, ignore_index=True)
names
```

```python
names.groupby('sex').births.sum()
```

```python
names['sex'].value_counts()
```

```python
total_births = names.pivot_table('births','year','sex',sum)
```

```python
total_births.tail()
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = (births / births.sum())
    return group
names = names.groupby(['year','sex']).apply(add_prop)
names
```

```python
names.tail()
```

```python
np.allclose(names.groupby(['year','sex']).prop.sum(), 1)
```

```python
grouped = names.groupby(['year','sex'])
```

```python
grouped
```

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]

top1000 = grouped.apply(get_top1000)
```

```python
top1000
```

```python
total_births = top1000.pivot_table('births', 'year', 'name', sum)
```

```python
total_births
```

```python
subset = total_births[['John','Harry','Mary','Marilyn']]
```

```python
subset
```

```python
subset.plot(subplots=True, figsize=(12,10),grid=False,title="Num of births per year")
```

```python
get_last_letter = lambda x: x[-1]
```

```python
last_letters = names.name.map(get_last_letter)
```

```python
last_letters.name = 'last_letter'
```

```python
table = names.pivot_table('births', last_letters,['sex','year'],sum)
```

```python
table
```

```python
subtable = table.reindex(columns=[1910,1960,2010],level='year')
```

```python
subtable
```

```python
subtable.sum()
```

```python
latter_prop = subtable / subtable.sum().astype(float)
```

```python
%timeit latter_prop
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,1)
latter_prop['M'].plot(kind='bar', rot=0,ax=axes[0], title='Male')
latter_prop['F'].plot(kind='bar', rot=0,ax=axes[1], title='Female', legend=False)
```
